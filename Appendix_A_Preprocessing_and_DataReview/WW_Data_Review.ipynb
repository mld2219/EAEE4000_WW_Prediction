# Understanding the Project Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
influent = pd.read_excel(r"C:\Users\18455\EAEE4000_ML_Jupyter\EAEE4000_WW_Prediction\Appendix_A_Preprocessing_and_DataReview\raw_influent.xlsx")
influent
```

```python
a = influent['TKN']

b = influent['SS']
b1 = influent['TS']
b2 = influent['VSS']
b3 = influent['VTS']
c1 = influent['BOD5']
c2 = influent['CBOD5']
d1 = influent['P-TOT']
d2 = influent['P-SOL']

plt.xlabel('Predictand - TKN')
plt.ylabel('Predictors - Influent Characteristics')
plt.scatter(a, b, label='SS')
plt.scatter(a, b1, label='TS')
plt.scatter(a, b2, label='VSS')
plt.scatter(a, b3, label='VTS')
plt.scatter(a, c1, label='BOD5')
plt.scatter(a, c2, label='CBOD5')
plt.scatter(a, d1, label='P-TOT')
plt.scatter(a, d2, label='P-SOL')

plt.legend()
print("Strong relationship between TKN and Other Influent Concentrations:")
```

```python
a = influent['Nitrates']

b = influent['SS']
b1 = influent['TS']
b2 = influent['VSS']
b3 = influent['VTS']
c1 = influent['BOD5']
c2 = influent['CBOD5']
d1 = influent['P-TOT']
d2 = influent['P-SOL']

plt.xlabel('Predictants - Nitrates')
plt.ylabel('Predictors - Influent Characteristics & Wet Weather')
plt.scatter(a, b, label='SS')
plt.scatter(a, b1, label='TS')
plt.scatter(a, b2, label='VSS')
plt.scatter(a, b3, label='VTS')
plt.scatter(a, c1, label='BOD5')
plt.scatter(a, c2, label='CBOD5')
plt.scatter(a, d1, label='P-TOT')
plt.scatter(a, d2, label='P-SOL')

plt.legend()
print("Weaker relationship between Nitrates and Other Influent Concentrations:")
```

```python
a = influent['Nitrates']

b = influent['TKN']

plt.xlabel('Nitrates')
plt.ylabel('TKN')
plt.scatter(a, b, label='TKN')

plt.legend()
print("Weaker relationship between Nitrates and TKN. Conclusion: TKN has stronger relationships in review of raw data.")
```
